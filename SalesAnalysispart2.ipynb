```python
df= all_data[all_data['Order ID'].duplicated(keep=False)]

df
```

```python
df['grouping']=df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
```

```python
df
```

```python
df=df[['Order ID','grouping']].drop_duplicates()
```

```python
df.head(50)
```

```python
#HAVE TO COUNT PAIR OF OFTEN SOLD ITEM 
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
```

```python
for row in df['grouping']:
    
    row_list = row.split(',')
    
    count.update(Counter(combinations(row_list,2)))
```

```python
count.most_common(10)
```

```python
product_group= all_data.groupby('Product')

quantities_ordered = product_group.sum()['Quantity Ordered']
```

```python
products=[product for product, df in product_group]
```

```python
plt.bar(products,quantities_ordered)

plt.xticks(products,rotation='vertical',size =8)

plt.ylabel('quantities_ordered')

plt.xlabel('product sold')

plt.show()
```

```python
Price =all_data.groupby('Product').mean()['Price Each']
```

```python
fig, ax1 = plt.subplots()


ax2 = ax1.twinx()
ax1.bar(products, quantities_ordered,color='g')
ax2.plot(products, Price, 'b-')

ax1.set_xlabel('Product Name ')
ax1.set_ylabel('quantities_ordered', color='g')
ax2.set_ylabel('Price', color='b')

ax1.set_xticklabels(products,rotation='vertical',size =8)

plt.show()
```
